2<sup>7</sup>=128 is the first power of two whose leading digits are "12".
The next power of two whose leading digits are "12" is 2<sup>80</sup>.

Define p(L,n) to be the nth-smallest value of j such that the base 10 representation of 2<sup>j</sup> begins with the digits of L.
So p(12,1)=7 and p(12,2)=80.

You are also given that p(123,45)=12710.

Find p(123,678910).

```python
num = 1
power = 0
count = 0

while count != 678910:
    num *= 2
    power += 1
    if str(num)[0:2] == "123":
        count += 1
    
print(power)
```
