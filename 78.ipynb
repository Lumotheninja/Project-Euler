Find the least value of n for which p(n) is divisible by one million.

```python
def pentagon(k):
    return int(k*(3*k-1)/2)
```

```python
from collections import defaultdict
import math

p= {}
x = 0
p[x] = 1


while p[x] % 1000000 != 0:
    x += 1
    p[x] = 0
    k = 1
    while x - pentagon(k) >= 0:
        p[x] += int((-1)**(k+1))*p[x-pentagon(k)]
        k += 1
    k = -1
    while x - pentagon(k) >= 0:
        p[x] += int((-1)**(k+1))*p[x-pentagon(k)]
        k -= 1
```
