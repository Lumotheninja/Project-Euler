In the card game poker, a hand consists of five cards and are ranked, from lowest to highest, in the following way:

High Card: Highest value card.
One Pair: Two cards of the same value.
Two Pairs: Two different pairs.
Three of a Kind: Three cards of the same value.
Straight: All cards are consecutive values.
Flush: All cards of the same suit.
Full House: Three of a kind and a pair.
Four of a Kind: Four cards of the same value.
Straight Flush: All cards are consecutive values of same suit.
Royal Flush: Ten, Jack, Queen, King, Ace, in same suit.
The cards are valued in the order:
2, 3, 4, 5, 6, 7, 8, 9, 10, Jack, Queen, King, Ace.

If two players have the same ranked hands then the rank made up of the highest value wins; for example, a pair of eights beats a pair of fives (see example 1 below). But if two ranks tie, for example, both players have a pair of queens, then highest cards in each hand are compared (see example 4 below); if the highest cards tie then the next highest cards are compared, and so on.

The file, poker.txt, contains one-thousand random hands dealt to two players. Each line of the file contains ten cards (separated by a single space): the first five are Player 1's cards and the last five are Player 2's cards. You can assume that all hands are valid (no invalid characters or repeated cards), each player's hand is in no specific order, and in each hand there is a clear winner.

How many hands does Player 1 win?

```python
class Card():
    
    num = 0
    suit = 0
    string = ""
    
    def __repr__(self):
        return self.string
    
    def __init__(self, card_str):
        self.string = card_str
        num = card_str[0]
        if num == "T":
            self.num = 10
        elif num == "J":
            self.num = 11
        elif num == "Q":
            self.num = 12
        elif num == "K":
            self.num = 13
        elif num == "A":
            self.num = 14
        else:
            self.num = int(num)
        
        suit = card_str[1]
        if suit == "S":
            self.suit = 4
        elif suit == "H":
            self.suit = 3
        elif suit == "C":
            self.suit = 2
        else:
            self.suit = 1
```

```python
from collections import defaultdict

def pokerhands(hand):
    
    # full combos, check suit before checking num
    # straight flush = 8, 1st dp is suit, 2nd-3rd dp is num 
    # four of a kind = 7, 1st-2nd dp is num
    # full house = 6, 1st-2nd dp is num
    # flush = 5, 1st dp is suit, 2nd-3rd dp is num
    # straight = 4, 1st dp is suit, 2nd-3rd dp is num
    
    # others, check num before checking suit
    # three of a kind = 3, 1st-2nd dp is num
    # 2 doubles = 2, 1st-2nd dp is num, 3rd dp is suit
    # 1 double = 1, 1st-2nd dp is num, 3rd dp is suit
    # single card = 0, 1st-2nd dp is num, 3rd dp is suit
    
    cards = [Card(x) for x in hand]
    cards.sort(key= lambda item: (item.num, item.suit))
    
    
    # counting straights
    straight_count = 0
    suit_count_dict = defaultdict(int)
    straight = False
    flush = False
    
    # counting fours, triples and doubles
    num_count_dict = defaultdict(int)
    fours_num = 0
    triples_num = 0
    doubles_count = 0
    doubles_num = 0
    doubles_suit = 0
    
    for i in range(len(cards)):
        straight_count += cards[i].num == 2 and cards[i-1].num == 14
        straight_count += cards[i].num - cards[i-1].num == 1
        if straight_count == 4:
            straight = True
        
        suit_count_dict[cards[i].suit] += 1
        if suit_count_dict[cards[i].suit] == 5:
            flush = True
            
        num_count_dict[cards[i].num] += 1
        if num_count_dict[cards[i].num] == 4:
            fours_num = cards[i].num
            triples_num = 0
        if num_count_dict[cards[i].num] == 3:
            triples_num = cards[i].num
            doubles_num = 0
            doubles_suit = 0
            doubles_count -= 1
        if num_count_dict[cards[i].num] == 2:
            doubles_num = cards[i].num
            doubles_suit = cards[i].suit
            doubles_count += 1
    
    top_num = cards[-1].num
    top_suit = cards[-1].suit
    
    if flush and straight:
        print ("straight flush")
        return 8 + 0.1 * top_suit + 0.001 * top_num
    if fours_num:
        print ("4 of a kind")
        return 7 + 0.01 * fours_num
    if triples_num and doubles_count == 1:
        print ("full house")
        return 6 + 0.01 * triples_num
    if flush:
        print ("flush")
        return 5 + 0.1 * top_suit + 0.001 * top_num
    if straight:
        print ("straight")
        return 4 + 0.1 * top_suit + 0.001 * top_num
    if triples_num:
        print ("triples")
        return 3 + 0.01 * triples_num
    if doubles_count == 2:
        print ("2 pairs")
        return 2 + 0.01 * doubles_num + 0.001 * doubles_suit
    if doubles_num:
        print ("1 pair")
        return 1 + 0.01 * doubles_num + 0.001 * doubles_suit
    print ("single")
    return 0 + 0.01 * top_num + 0.001 * top_suit
        
```

```python
p1win = 0

with open("p054_poker.txt") as file:
    for lines in file.readlines():
        p1score = pokerhands(lines.strip("\n").split(" ")[0:5])
        p2score = pokerhands(lines.strip("\n").split(" ")[5:])
        p1win += p1score > p2score
```
