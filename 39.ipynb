If p is the perimeter of a right angle triangle with integral length sides, {a,b,c}, there are exactly three solutions for p = 120.

{20,48,52}, {24,45,51}, {30,40,50}

For which value of p â‰¤ 1000, is the number of solutions maximised?

```python
def gcd(x, y): 
  
   while(y): 
       x, y = y, x % y 
  
   return x 
```

```python
from collections import defaultdict

pythagoras_dict = defaultdict(int)

for k in range(1, 100):
    for m in range(1,10000):
        for n in range(1, m):
            if gcd(m, n) == 1:
                if m%2 ==0 or n%2 ==0:
                    sum_abc = k*2*(m**2 + m*n)
                    if sum_abc <= 1000:
                        pythagoras_dict[sum_abc] += 1
                    else:
                        break
```

```python
max_val = 0
max_arg = 0

for key in pythagoras_dict.keys():
    print(key, pythagoras_dict[key])
    if pythagoras_dict[key] > max_val:
        max_val = pythagoras_dict[key]
        max_arg = key
```
