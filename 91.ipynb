```python
import math

count = 0
lim = 50
# i is x-coordinate
# j is y-coordinate
# k is range from 1 to i inclusive
# l is range from 1 to j inclusive
for i in range(1,lim+1):
    for j in range(1,lim+1):
        for k in range(0, i+1):
            for l in range(0, j+1):
                a = (k, j)
                b = (i, l)
                if a[0] == 0 and b[1] == 0:
                    count += 1
                    #print(a,b)
                elif a == (i, j) and b == (i, 0):
                    count += 1
                    #print(a,b)
                elif a == (0, j) and b == (i, j):
                    count += 1
                    #print(a,b)
                elif a == b:
                    continue
                else:
                    sideoa = a[0]**2 + a[1]**2
                    sideob = b[0]**2 + b[1]**2
                    sideab = (a[0] - b[0])**2 + (a[1] - b[1])**2
                    hypothenus = max(sideoa, sideob, sideab)
                    if hypothenus - sideoa - sideob - sideab == -hypothenus:
                        count += 1
                        print ("h", a,b,sideoa, sideob, sideab)
count
```
