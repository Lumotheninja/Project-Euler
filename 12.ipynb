The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be: 1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ... Let us list the factors of the first seven triangle numbers:  
1: 1  
3: 1,3  
6: 1,2,3,6  
10: 1,2,5,10  
15: 1,3,5,15  
21: 1,3,7,21  
28: 1,2,4,7,14,28  
We can see that 28 is the first triangle number to have over five divisors.  
What is the value of the first triangle number to have over five hundred divisors?

```python
def apsum(x):
    return (x/2)*(1+x)
```

```python
import math
def divisor(x):
    count=0
    for y in range(1,math.floor(math.sqrt(x)+1)):
        if x%y == 0:
            if math.sqrt(x) == float(y):
                count+=1
            else:
                count+=2
    return count
```

```python
def find_500_divisor():
    max=0
    x=1
    while (max <500):
        div = divisor(apsum(x))
        if div > max:
            max=div
        x+=1
    return apsum(x-1)
```
